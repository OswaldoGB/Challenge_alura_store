### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



```python
facturacion_tienda1 = tienda1["Precio"].sum()
facturacion_tienda2 = tienda2["Precio"].sum()
facturacion_tienda3 = tienda3["Precio"].sum()
facturacion_tienda4 = tienda4["Precio"].sum()

print("Facturación total por tienda:\n")

print(f"Tienda 1: ${facturacion_tienda1:,.2f}")
print(f"Tienda 2: ${facturacion_tienda2:,.2f}")
print(f"Tienda 3: ${facturacion_tienda3:,.2f}")
print(f"Tienda 4: ${facturacion_tienda4:,.2f}")
```

# 2. Ventas por categoría

```python
print("="*50)
print("VENTAS POR CATEGORÍA")
print("="*50)

def ventas_por_categoria(tienda, nombre_tienda):
    
    categorias = tienda["Categoría del Producto"].value_counts()
    total_ventas = categorias.sum()
    
    print(f"\n📍 {nombre_tienda}")
    print("-"*40)
    
    for categoria, cantidad in categorias.items():
        porcentaje = (cantidad / total_ventas) * 100
        print(f"{categoria:<25} {cantidad:>8,} ventas   ({porcentaje:5.2f}%)")

ventas_por_categoria(tienda1, "Tienda 1")
ventas_por_categoria(tienda2, "Tienda 2")
ventas_por_categoria(tienda3, "Tienda 3")
ventas_por_categoria(tienda4, "Tienda 4")
```

# 3. Calificación promedio de la tienda


```python
print("Promedio de calificación de clientes:\n")

print(f"Tienda 1: {tienda1['Calificación'].mean():.2f}")
print(f"Tienda 2: {tienda2['Calificación'].mean():.2f}")
print(f"Tienda 3: {tienda3['Calificación'].mean():.2f}")
print(f"Tienda 4: {tienda4['Calificación'].mean():.2f}")
```

# 4. Productos más y menos vendidos

```python
print("Productos más y menos vendidos:\n")

print("Tienda 1")
print("Más vendido:", tienda1["Producto"].value_counts().idxmax())
print("Menos vendido:", tienda1["Producto"].value_counts().idxmin())

print("\nTienda 2")
print("Más vendido:", tienda2["Producto"].value_counts().idxmax())
print("Menos vendido:", tienda2["Producto"].value_counts().idxmin())

print("\nTienda 3")
print("Más vendido:", tienda3["Producto"].value_counts().idxmax())
print("Menos vendido:", tienda3["Producto"].value_counts().idxmin())

print("\nTienda 4")
print("Más vendido:", tienda4["Producto"].value_counts().idxmax())
print("Menos vendido:", tienda4["Producto"].value_counts().idxmin())
```

# 6. Gráfico

```python
import matplotlib.pyplot as plt

def grafico_categorias(tienda, nombre_tienda):

    categorias = tienda["Categoría del Producto"].value_counts()

    plt.figure(figsize=(10,6))

    barras = plt.bar(
        categorias.index,
        categorias.values
    )

    plt.title(f"Ventas por Categoría - {nombre_tienda}", fontsize=16, fontweight="bold")
    plt.xlabel("Categoría", fontsize=12)
    plt.ylabel("Cantidad de Ventas", fontsize=12)

    plt.xticks(rotation=30)
    plt.grid(axis="y", linestyle="--", alpha=0.4)

    # Etiquetas de valor encima de cada barra
    for barra in barras:
        altura = barra.get_height()
        plt.text(
            barra.get_x() + barra.get_width()/2,
            altura,
            f"{int(altura):,}",
            ha="center",
            va="bottom",
            fontsize=10
        )

    plt.tight_layout()
    plt.show()


grafico_categorias(tienda1, "Tienda 1")
grafico_categorias(tienda2, "Tienda 2")
grafico_categorias(tienda3, "Tienda 3")
grafico_categorias(tienda4, "Tienda 4")
```
